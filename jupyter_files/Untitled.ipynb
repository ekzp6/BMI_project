```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_excel('raw_data-xlsx/merged_data.xlsx') 
samples = pd.read_excel('sample_annotation.xlsx')
```

```python
df1 = raw_data[raw_data['Source_file']==1]
df2 = raw_data[raw_data['Source_file']==2]
df3 = raw_data[raw_data['Source_file']==3]
df4 = raw_data[raw_data['Source_file']==4]
```

```python
samples1 = samples[samples['File']==1]
samples2 = samples[samples['File']==2]
samples3 = samples[samples['File']==3]
samples4 = samples[samples['File']==3]
```

```python
KEEP = ['Source_file','Block', 'Column', 'Row', 'Name', 'ID', 'F532 Median',
       'B532 Median', 'F635 Median', 'B635 Median', 'F532 Median - B532',
       'F635 Median - B635', 'F532 Total Intensity', 'F635 Total Intensity', 'SNR 532', 'SNR 635']
```

```python
data = [df1,df2,df3,df4]
lables = [samples1,samples2,samples3,samples4]
```

```python
for i in range(len(data)):
    data[i] =  pd.merge(data[i], lables[i], how='left', on =['Block'])
```

```python
pd.concat(data).to_excel('Preprocessed.xlsx')
```
